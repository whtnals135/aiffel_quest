```python
import random as r

# 무작위 실행 함수
def rand_choose():
    random_value = r.random()
    if random_value < 0.25:
        random_value = 1

    elif random_value < 0.5:
        random_value = 2

    elif random_value < 0.75:
        random_value = 3

    else:
        random_value = 4

    return random_value


# 아이펠 클래스
class Aiffel:

    core = 'core'
    number = 7
    year = 2023

    # 오자마자한 자기소개
    def introducing(self):
        print(f'{Aiffel.year}에 입학한 {Aiffel.core}과정에 {Aiffel.number}기입니다!')

    # 하루 일과
    def LMS(self):
        print('LMS 수업하기')

    def quest(self):
        print('Quest 진행하기')

    def cart(self):
        print('카트타기')

    def tetris(self):

        print('테트리스 하기')
        random_value = rand_choose()

        if(random_value <= 2):
            print('결과 :광석님을 쉽게 이겨버림')
        else:
            print('결과 : 너무 아쉽게 광석님께 저버림')


# 그루 클래스
class grew(Aiffel):

    def __init__(self, name, blood, mbti):
        self.name = name
        self.blood = blood
        self.mbti = mbti

    def meal(self):
        print('밥먹기')

    def sleep(self):
        print('잠자기 zzz')


student1 = grew('조보겸', 24, 'ISTP')

# 입학하자마자한 자기소개
student1.introducing()

# 시작하는 하루일과
print()
print('아이펠 하루 일과!')
print('-'*20)
for i in range(3):
    random_value = rand_choose()
    if random_value == 1:
        student1.LMS()

    elif random_value == 2:
        student1.quest()

    elif random_value == 3:
        student1.cart()

    else:
        student1.tetris()

print('-'*20)
print()

student1.meal()
student1.sleep()

print()
print('하루 끝!')
```
